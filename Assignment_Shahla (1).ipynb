# Q1. Arithmetic and Assignment Operators



```python
# Creating two variables
a = 10   # a ki value 10
b = 3    # b ki value 3

# Arithmetic operations
print("Addition:", a + b)          # 10 + 3
print("Subtraction:", a - b)       # 10 - 3
print("Multiplication:", a * b)    # 10 * 3
print("Division:", a / b)          # 10 / 3 normal division
print("Floor Division:", a // b)   # 10 // 3 result without decimal
print("Modulus:", a % b)           # 10 % 3 remainder
print("Exponentiation:", a ** b)   # 10^3 (10 power 3)

# Assignment operators (a ki value change hoti hai)
a += 5  # a = a + 5
print("After a += 5:", a)

a -= 2  # a = a - 2
print("After a -= 2:", a)

a *= 3  # a = a * 3
print("After a *= 3:", a)

a /= 2  # a = a / 2
print("After a /= 2:", a)
```

# Q2. Comparison and Logical Operators


```python
# Taking input from user
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

# Comparison operators
print("x == y:", x == y)   # Dono same?
print("x != y:", x != y)   # Dono different?
print("x > y:", x > y)     # x bada?
print("x < y:", x < y)     # x chota?
print("x >= y:", x >= y)   # x bada ya barabar?
print("x <= y:", x <= y)   # x chota ya barabar?

# Logical operators
print("x > 5 and y > 5:", x > 5 and y > 5)  # Dono 5 se bade?
print("x > 5 or y > 5:", x > 5 or y > 5)    # Koi ek 5 se bara?
print("not(x > y):", not (x > y))           # x > y ko ulta
```

# Q3. Bitwise Operators

```python
x = 5
y = 3

print("Bitwise AND:", x & y)      # AND
print("Bitwise OR:", x | y)       # OR
print("Bitwise XOR:", x ^ y)      # XOR
print("Bitwise NOT (~x):", ~x)    # NOT
print("Left Shift (x << 1):", x << 1)   # Multiply by 2
print("Right Shift (x >> 1):", x >> 1)  # Divide by 2
```

# Q4.Identity & Membership Operators

```python
numbers = [1, 2, 3, 4, 5]

a = numbers       # Same list
b = numbers       # Same list
c = numbers.copy()  # New copy (different)

# Identity operators
print("a is b:", a is b)       # Same memory?
print("a is not c:", a is not c)   # Different memory?

# Membership operators
print(3 in numbers)            # 3 list me?
print(10 not in numbers)       # 10 list me nahi?
```

# Q5.String Operations

```python
sentence = "Python programming is very easy."

print("Length:", len(sentence))                  # Total characters
print("Uppercase:", sentence.upper())            # Bada text
print("Lowercase:", sentence.lower())            # Chota text
print("Is 'easy' in sentence:", "easy" in sentence)  # Word search
print("Replace 'easy' with 'fun':", sentence.replace("easy", "fun"))  # Replace
print("Slice (0 to 10):", sentence[0:10])        # First 10 characters
print("Reverse string:", sentence[::-1])         # Full reverse
```
